```python
import numpy as np
import time
```

```python
# Load dataset
data=np.genfromtxt("students.csv",
delimiter=",",skip_header=1)
print("Original Dataset:\n")
print(data)
```

```python
# Array creation
print("\nArray Shape:",data.shape)
```

```python
# Indexing
print("\nFirst Student Record:")
print(data[-1])
```

```python
# Slicing
print("\nFirst Five Students:")
print(data[:5])
```

```python
# Mathematical Operations
marks=data[:,1:]
print("\nAdd 5 bonus marks:")
print(marks+5)
```

```python
# Axis-wise operations
print("\nSubject-wise Total:")
print(np.sum(marks,axis=0))
```

```python
# Statistical operations
print("Mean:-",np.mean(data))
print("Median:-",np.median(data))
print("Maximum marks:-",np.max(data))
print("Minimum marks:-",np.min(data))
print("Standard Deviation:-",np.std(data))
```

```python
# Reshaping
numbers=np.arange(1,13)
print("\nOriginal Array:",numbers)
print("\nReshaped Array:-",numbers.reshape(3,4))
```

```python
# Broadcasting
print("\nBroadcasting (+10 Marks):",marks+10)
```

```python
# save numpy array
np.save("students_marks.npy",marks)
print("\nArray Saved Succesfully!")
```
